### Import Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.dpi'] = 144
from scipy.stats import norm
```

### 3D Gaussian in Cartesian Coordinate

```python
# Mean and covariance of the distribution
mean = [0, 0]
cov = [[1, 0], [0, 1]]

# Generate grid points
x = np.linspace(-2, 0.5, 30)
y = np.linspace(-3, 3, 30)
x, y = np.meshgrid(x, y)

# Calculate the Gaussian distribution on the grid
z = np.exp(-((x - mean[0])**2 + (y - mean[1])**2) / 2)

# Create a 3D plot with a specific aspect ratio
fig = plt.figure(figsize=(12, 8))  # Adjust the figure size
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z, cmap='viridis')

ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Probability Density')

# Set the aspect of the plot.
ax.set_box_aspect([1,1,0.3])  # Adjust the scale of the axes

plt.show()
```

### 3D Gaussian in Polar Coordinate

```python
# Parameters for the bivariate normal distribution
sigma2 = 1  # Variance

# Create a grid in polar coordinates
r = np.linspace(0, 3, 100)
theta = np.linspace(np.pi/2, 3*np.pi/2, 100)
R, Theta = np.meshgrid(r, theta)

# Convert polar coordinates to Cartesian coordinates
X = R * np.cos(Theta)
Y = R * np.sin(Theta)

# Calculate the probability density using the simplified PDF equation
Z = (1 / (2 * np.pi * sigma2)) * np.exp(-0.5 * (X**2 + Y**2) / sigma2)

# Create a 3D plot in polar coordinates
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(R * np.cos(Theta), R * np.sin(Theta), Z, cmap='viridis')

# Set the aspect of the plot and hide axis labels
ax.set_box_aspect([1,1,0.3])

ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Probability Density')
ax.set_title('3D Plot of Bivariate Normal Distribution in Polar Coordinates')

plt.show()
```

### 1. Normal Distribution Graphs as a Function of Distance and Speed, R(d,v)

**Speed 1**

```python
# Define the range of distances from the equipment you want to model, from 0 to 10 meters
distances = np.linspace(0, 10, 1000)

# In this case the variable which could be governed by speed is the standard deviation
std_dev = 4  # Standard deviation = kv

# Calculate the half-normal distribution for the risk probability density
risk_pdf = norm.pdf(-distances, 0, std_dev)

# Plotting the risk probability density function
plt.plot(distances, risk_pdf, label='Risk Probability Density Function')
plt.xlabel('Distance from Equipment (meters)')
plt.ylabel('Risk Probability Density')
plt.title('Risk PDF with Respect to Distance from Equipment')
plt.legend()
plt.grid(True)
plt.show()
```

### 2. Normal Distribution Graphs as a Function of  and Speed, R(,v)

**Speed 1**

```python
# Define the range of angles from -90 to 90 degrees
angles = np.linspace(-90, 90, 1000)

# In this case the variable which could be governed by speed is the standard deviation
std_dev_angle = 40  # Standard deviation = kv

# Calculate the probability density function for risk based on angle
angle_risk_pdf = norm.pdf(angles, 0, std_dev_angle)

# Plotting the risk probability density function with respect to angle
plt.plot(angles, angle_risk_pdf, label='Risk Probability Density Function')
plt.xlabel('Angle (degrees)')
plt.ylabel('Risk Probability Density')
plt.title('Risk PDF with Respect to Angle from Equipment')
plt.axvline(0, color='red', linestyle='--', label='Highest Risk (0)')
plt.legend()
plt.grid(True)
plt.show()
```
