```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
plt.rcParams['figure.dpi'] = 144

# Mean and covariance of the distribution
mean = [0, 0]
cov = [[1, 0], [0, 1]]

# Generate grid points
x = np.linspace(-2, 0.5, 30)
y = np.linspace(-3, 3, 30)
x, y = np.meshgrid(x, y)

# Calculate the Gaussian distribution on the grid
z = np.exp(-((x - mean[0])**2 + (y - mean[1])**2) / 2)

# Create a 3D plot with a specific aspect ratio
fig = plt.figure(figsize=(12, 8))  # Adjust the figure size
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z, cmap='viridis')

# Set the aspect of the plot. 'auto', 'equal', or a number.
ax.set_box_aspect([1,1,0.3])  # Adjust the scale of the axes

ax.set_xticks([])  # Hide x-axis labels
ax.set_yticks([])  # Hide y-axis labels
ax.set_zticks([])  # Hide z-axis labels

plt.show()
```
